```python
import glob
import os
import pandas as pd
```

```python
folder_name = '/home/sai/Projects/COVID-19/csse_covid_19_data/'
file_type = 'csv'
seperator =','
#df = pd.concat([pd.read_csv(f, sep=seperator) for f in glob.glob(folder_name + "/*."+file_type)],ignore_index=True)
```

```python
file_list = glob.glob(folder_name + "csse_covid_19_daily_reports_us" +"/*."+file_type)
df_list = []
for f in file_list:
    tmp_df = pd.read_csv(f, sep=',')
    tmp_df['date'] = os.path.splitext(os.path.basename(f))[0]
    df_list.append(tmp_df)
df = pd.concat(df_list)
```

```python
df_sorted = df.sort_values(["Province_State", "date"])
```
