```python
file_list = glob.glob(folder_name + "csse_covid_19_daily_reports" +"/*."+file_type)
df_list = []
for f in file_list:
    tmp_df = pd.read_csv(f, sep=',')
    tmp_df['date'] = os.path.splitext(os.path.basename(f))[0]
    df_list.append(tmp_df)
df_daily = pd.concat(df_list)
```

```python
df_daily.isna().sum()
```

```python
df_daily['Province_State'] = df_daily['Province_State'].combine_first(df_daily['Province/State'])
df_daily['Country_Region'] = df_daily['Country_Region'].combine_first(df_daily['Country/Region'])
df_daily['Last_Update'] = df_daily['Last_Update'].combine_first(df_daily['Last Update'])
df_daily['Lat'] = df_daily['Lat'].combine_first(df_daily['Latitude'])
df_daily['Long_'] = df_daily['Long_'].combine_first(df_daily['Longitude'])
df_daily.drop(columns=['Province/State', 'Country/Region', 'Last Update', 'Latitude', 'Longitude'], inplace=True)
```

```python
df_daily_sorted = df_daily.sort_values(["Country_Region", "Province_State", "Admin2"])
```
